```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.figure(figsize = (14, 10), dpi = 80)

x_axis = np.arange(start=0, stop=130)

read_tsv = pd.read_csv('./Findings_tsv/Decision_Tree_RGB_-_Ark1.tsv', sep='\t',  header=None).to_numpy().flatten()
read_tsv_1 = pd.read_csv('./Findings_tsv/MLPC_Grey_-_Ark1.tsv', sep='\t',  header=None).to_numpy().flatten()
read_tsv_2 = pd.read_csv('./Findings_tsv/MLPC_RGB_-_Ark1.tsv', sep='\t',  header=None).to_numpy().flatten()
read_tsv_3 = pd.read_csv('./Findings_tsv/Random_Forest_RGB_-_Ark1.tsv', sep='\t',  header=None).to_numpy().flatten()

Decision_Tree_RGB = (np.sort(read_tsv))
MLPC_Grey = (np.sort(read_tsv_1))
MLPC_RGB = (np.sort(read_tsv_2))
Random_Forest_RGB = (np.sort(read_tsv_3))

plt.ylabel('Youdens Index')
plt.xlabel('Results')

plt.scatter(x_axis, Decision_Tree_RGB, s=20, color='red')
plt.show()
```

```python
pd.set_option('display.max_rows', 500)
youdens_score = np.arange(start=0.00, stop=1.01, step=0.01) 


# DECISION TREE
unique_dc_tree, counts_dc_tree = np.unique(Decision_Tree_RGB, return_counts=True)
dc_tree_unique_counts = dict(zip(unique_dc_tree, counts_dc_tree))

dec_tree_construct = []
for n in youdens_score:
    dec_tree_construct.append(0.00)

for count, n in enumerate(youdens_score):
    for k in dc_tree_unique_counts:
        n_round = np.round(n, 2)
        k_stripped = float(k.replace(",", "."))

        if k_stripped == n_round:
            dec_tree_construct[count]=dc_tree_unique_counts[k]

# MLPC_Grey
unique_mlpc_grey, counts_mlpc_grey = np.unique(MLPC_Grey, return_counts=True)
mlpc_grey_unique_counts = dict(zip(unique_mlpc_grey, counts_mlpc_grey))

mlpc_grey_construct = []
for n in youdens_score:
    mlpc_grey_construct.append(0.00)

for count, n in enumerate(youdens_score):
    for k in mlpc_grey_unique_counts:
        n_round = np.round(n, 2)
        k_stripped = float(k.replace(",", "."))

        if k_stripped == n_round:
            mlpc_grey_construct[count]=mlpc_grey_unique_counts[k]
            
# MLPC_RGB
unique_mlpc_rgb, counts_mlpc_rgb = np.unique(MLPC_RGB, return_counts=True)
mlpc_rgb_unique_counts = dict(zip(unique_mlpc_rgb, counts_mlpc_rgb))

mlpc_rgb_construct = []
for n in youdens_score:
    mlpc_rgb_construct.append(0.00)

for count, n in enumerate(youdens_score):
    for k in mlpc_rgb_unique_counts:
        n_round = np.round(n, 2)
        k_stripped = float(k.replace(",", "."))

        if k_stripped == n_round:
            mlpc_rgb_construct[count]=mlpc_rgb_unique_counts[k]
                       
        
# Random_Forest_RGB
unique_rndm_frst, counts_rndm_frst = np.unique(Random_Forest_RGB, return_counts=True)
rndm_frst_unique_counts = dict(zip(unique_rndm_frst, counts_rndm_frst))

rndm_frst_construct = []
for n in youdens_score:
    rndm_frst_construct.append(0.00)

for count, n in enumerate(youdens_score):
    for k in rndm_frst_unique_counts:
        n_round = np.round(n, 2)
        k_stripped = float(k.replace(",", "."))

        if k_stripped == n_round:
            rndm_frst_construct[count]=rndm_frst_unique_counts[k]
                       
                
                
data = {'Youdens_index': np.arange(start=0.00, stop=1.01, step=0.01).tolist(),
       'Decision_tree_count':dec_tree_construct,
       'MLPC_grey_count':mlpc_grey_construct,
       'MLPC_RGB_count':mlpc_rgb_construct,
       'Random_forest_count': rndm_frst_construct
       }
    
 
# Create DataFrame  
df = pd.DataFrame(data)  
  
# Print the output.  
print(df)  
```
